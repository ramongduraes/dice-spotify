```python
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
print(df.shape)
df.head()
```

```python
# Checking for null values
assert len(df[df.isnull().any(axis=1)]) == 0 

# Dropping the ID column 
df.drop(columns=['id'], inplace=True)

# Transforming categorical variables
df['explicit'] = df['explicit'].astype('category')

# Transforming release date
df['release_date'] = pd.to_datetime(df['release_date'], infer_datetime_format=True)

# Keeping only the first artist of the list
df.artists = df.artists.apply(lambda x: x.replace("\"","").replace("\'", "").replace("[", "").replace("]", "").split(",")[0])

# Picking recent songs only
threshold_year = 1980
df = df[df.year > threshold_year]
print(df.shape)
```

## Scaling

```python
from sklearn import preprocessing
```

```python
df.head()
```

```python
df.set_index(['artists'], inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
numeric_df = df.drop(columns=['artists', 'name', 'release_date', 'year', 'key', 'mode'])
```

```python
minmax_processed = preprocessing.MinMaxScaler().fit_transform(numeric_df)
```

```python
df_numeric_scaled = pd.DataFrame(minmax_processed, index=numeric_df.index, columns=numeric_df.columns)
```

___

```python
Nc = range(1, 10)
kmeans = [KMeans(n_clusters=i) for i in Nc]

score = [kmeans[i].fit(df_numeric_scaled).score(df_numeric_scaled) for i in range(len(kmeans))]

import matplotlib.pyplot as plt

plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()

from kneed import KneeLocator

kneedle = KneeLocator(Nc, score, S=1.0, curve="concave", direction="increasing")

C = kneedle.knee

kmeans = KMeans(n_clusters=C)
kmeans.fit(df_numeric_scaled)

len(kmeans.labels_)

df['cluster'] = kmeans.labels_

df.head()

df.groupby(['cluster']).count()['year'].values
```

```python
plt.figure(figsize=(12,7))
axis = sns.barplot(x=np.arange(0,7,1),y=df.groupby(['cluster']).count()['year'].values)
x=axis.set_xlabel("Cluster Number")
x=axis.set_ylabel("Number of songs")
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(numeric_df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
np.sum(df.cluster.values==5)
```

```python
len(df.cluster==2)
```

```python
finalDf = pd.concat([principalDf, df[['cluster']]], axis = 1)
```

```python
selected_clusters = [4, 6]

sns.scatterplot(
    "principal component 1",
    "principal component 2",
    data = finalDf[[c in selected_clusters for c in finalDf.cluster]].sample(100),
    hue='cluster',
    palette="Set2"
)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(numeric_df)
```

```python
finalDf = pd.concat([principalDf, df[['cluster']]], axis = 1)
```

```python
numeric_df['tsne-2d-one'] = tsne_results[:,0]
numeric_df['tsne-2d-two'] = tsne_results[:,1]
numeric_df['cluster'] = df['cluster']
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="cluster",
    palette=sns.color_palette("hls", 7),
    data=numeric_df,
    legend="full",
    alpha=0.3
)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
